## <font color='purple'> __EXPERIMENTO__: </font>

### Extraeremos la letra 'H' sin conocer previamente la ubicación exacta.

Intentaremos extraer la letra 'H' sin saber cuál es la ubicación exacta dónde se encuentra, para lo cual necesitamos llamarla tal como es "H"

````python
#primero guardo la última línea

ult_linea = exam_lines[-1]
print(ult_linea)
````

````python
# busco la posición de "H", que es la posición 14, tal como habíamos contado previamente
posicion_H = ult_linea.find('H')
print(posicion_H)
````

````python
# le pido que la imprima
letra_H = ult_linea[posicion_H]
print(letra_H)
````

Probaremos un método más corto

````python
#también podía hacerlo todo en una sola línea
# en este caso lo imprimirá como cadena de texto
letra_H2 = exam_lines[-1][exam_lines[-1].find('H')]
letra_H2
````

## <font color='green'>Actividad 8:</font>
Qué tipos de datos devuelven las siguientes líneas de código:

     1.) 2 / 3
     2.) 2 + 2.0
     3.) 1 + 1
     4.) "2" + "2"
     5.) 1 > 2
    
Tip:
1. Intenta responder esto mentalmente antes de verificar ejecutando el código en Python.
2. Utiliza la función `type()`

````python
# Tu código aquí ...
#descucbrimos que al usar .__name__ después de type, el resultado pasa de ser: <class 'float'> a simplemente: float
#lo que hace que el resultado se vea más armónico
a = 2 / 3
b = 2 + 2.0
c = 1 + 1
d = "2" + "2"
e = 1 > 2

print(f"El tipo de dato de 2 / 3 es: {type(a).__name__}, y su resultado es {(a)}")
print(f"El tipo de dato de 2 + 2.0 es: {type(b).__name__}, y su resultado es {(b)}")
print(f"El tipo de dato de 1 + 1 es: {type(c).__name__}, y su resultado es {(c)}")
print(f"El tipo de dato de '2' + '2' es: {type(d).__name__}, y su resultado es {(d)}")
print(f"El tipo de dato de 1 > 2 es: {type(e).__name__}, y su resultado es {(e)}")
````
