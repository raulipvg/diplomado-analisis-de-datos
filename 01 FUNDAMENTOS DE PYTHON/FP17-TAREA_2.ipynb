````python
# Tu código aquí ...
while True:
    p2=input('Ingrese un número par:')
    if not p2.isdigit():
       print(f"Erorr. {p2} no es un número")
       break
    else:
        p= int(p2)
        if p%2 == 0:
            print(f"El número ingresado es par: {p}")
            break
        else: print(f"{p} No es par, intente de nuevo.")
        continue
````

## <font color='green'>Actividad 7: Super challenging</font>

Eres el administrador de una tienda en línea. Cada producto en la tienda tiene un identificador único y un precio. Los datos de los productos se almacenan en una lista de diccionarios, donde cada diccionario representa un producto y contiene dos claves: "id" para el identificador y "precio" para el precio del producto.

Debes encontrar el primer producto cuyo precio sea mayor que el precio medio de todos los productos y que no haya sido visitado por ningún cliente antes. Los productos visitados están almacenados en un conjunto llamado "productos_visitados". Si todos los productos han sido visitados o el precio de un producto no visitado no es mayor que el precio medio, debes imprimir "No hay productos que cumplan con las condiciones".

````python
productos = [
    {"id": 1, "precio": 100},
    {"id": 2, "precio": 200},
    {"id": 5, "precio": 500},
    {"id": 3, "precio": 300},
    {"id": 4, "precio": 400},
    {"id": 6, "precio": 80}
]

productos_visitados = {1, 2, 4}

# Tu código aquí ...

# Inicializando variables
precio_medio = 0
k= 0
sum_precio = 0
producto_ok = None

#Cálculo de precio promedio y re-ordenamiento de los productos según id
for a in productos:
    sum_precio += a["precio"]
    k += 1
precio_medio = sum_precio/k

print(f"El precio medio es {precio_medio:.1f}")

#Búsqueda de primer producto que cumpla con las condiciones

for c in productos:
    if (c["id"] not in productos_visitados) and (c["precio"] > precio_medio):
        producto_ok = c
        break

#Impresión de resultados
if producto_ok:
    print(f"El primer producto de la lista, no visitado, y de precio mayor al promedio, es el producto {producto_ok['id']} y su precio es {producto_ok['precio']}")
else:
    print(f"No hay productos que cumplan con las condiciones")
````

## <font color='purple'> __EXPERIMENTO__: </font>

### Variación de Actividad 7, incorporando criterio de ordenación de Id de productos.
En la redacción del problema 7, se pide "encontrar **el primer producto** cuyo precio sea mayor que el precio medio de todos los productos y que no haya sido visitado por ningún cliente antes."

En la solución, el programa recorre la lista `productos`, cuyos elementos se encuentran naturalmente en el siguiente orden: `1 - 2 - 5 - 3 - 4 - 6`.
En esta configuración, el primer producto que cumple con las condiciones es el Id 5.

A continuación, se construye una variante del problema, en que antes de evaluar las condiciones, se revisan los productos de manera ordenada de acuerdo a su Id, es decir: `1 - 2 - 3 - 4 - 5 - 6`.

Para ello, se incorporan al código los siguientes elementos:

*   Creación de una lista con los valores Id de cada producto.
*   Ordenar esta lista usando el método `sort`.
*   Ejecutar la rutina de búsqueda del producto, siguiendo el orden de los Id.

En esta nueva configuración, la respuesta al problema se modifica, y el primero producto que cumple con las condición es el Id 3.

El código modificado se presenta en la siguiente celda:

````python
productos = [
    {"id": 1, "precio": 100},
    {"id": 2, "precio": 200},
    {"id": 5, "precio": 500},
    {"id": 3, "precio": 300},
    {"id": 4, "precio": 400},
    {"id": 6, "precio": 80}
]

productos_visitados = {1, 2, 4}

# Tu código aquí ...

# Inicializando variables
precio_medio = 0
k= 0
sum_precio = 0
producto_ok = None
lista_id = []

#Cálculo de precio promedio y re-ordenamiento de los productos según id
for a in productos:
    sum_precio += a["precio"]
    lista_id.append(a['id'])
    k += 1
precio_medio = sum_precio/k
lista_id.sort()
print(f"El precio medio es {precio_medio:.1f}")
print(f"La lista re-ordenada de productos es {lista_id}")

#Búsqueda de primer producto que cumpla con las condiciones
for b in lista_id:
    for c in productos:
        if (c["id"] == b) and (c["id"] not in productos_visitados) and (c["precio"] > precio_medio):
            producto_ok = c
            break
    if producto_ok:
          break

#Impresión de resultados
if producto_ok:
    print(f"El primer producto de la lista ordenada, no visitado, y de precio mayor al promedio es el producto {producto_ok['id']} y su precio es {producto_ok['precio']}")
else:
    print(f"No hay productos que cumplan con las condiciones")
````
