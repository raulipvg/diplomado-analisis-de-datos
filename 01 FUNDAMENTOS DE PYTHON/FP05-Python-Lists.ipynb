## <font color='blue'>**Copiando listas**</font>

Ten en cuenta que cuando copias una lista no haces una copia *profunda*.

````python
l = [1, 2, 3]
n = l
print(n)
````

````python
# Modificamos la lista l

l.append(4)
print(l)
````

````python
# Miren lo que ocurre con la lista n

print(n)
````

En el fondo, los nombres de listas ***l*** y ***n*** apuntan al mismo objeto en memoria.<br>
Comprobemos esto con la función `id()`.

````python
id(l) == id(n)
````

````python
# La solución es copiar con la función deepcopy()
# Para ello debemos cargar una librería llamada 'copy' la cual tiene la función 'deepcopy()'

import copy
m = copy.deepcopy(l)
````

````python
l == n == m
````

````python
id(l) == id(n)
````

````python
id(l) == id(m)
````

````python
l.append(5)
print(l)
print(n)
print(m)
print(f'{l == n}')
print(f'{l == m}')
````

## <font color='purple'> __EXPERIMENTO__: </font>

### Manejo de Listas
- Creamos una lista con los elementos: 1,'Hola', 'H', True, False, 5.5, 5 (diferentes tipos de datos)

- Luego Creamos una lista 2 con los elemento: 5,6,7

- Insertamos en la lista, en la posicion 1 la lista 2

- Contamos la cantidad de 5 que existen en la lista y observamos su comportamiento

- Contamos la cantidad de H que existen en la lista y posteriormente la eliminamos

- Elimina la H y posteriomente invertimos la lista

- Finalmente, la imprimimos

````python
lista = [ 1,'Hola', 'H', True, False, 5.5, 5]

lista_2 = [5,6,7]

lista.insert(1,lista_2)

#Observamos que solo cuenta el 5 de la lista y no el anidado
print(f'cantidad de 5: {lista.count(5)}')
#Observamos que cuenta la H
print(f'cantidad de H: {lista.count("H")}')

lista.remove('H')
lista.reverse()
print(lista)
````
