## <font color='green'>Tarea 1: Challenging</font>
Cree una clase Esfera que acepte un radio en la instanciación y tenga  métodos de volumen y área de superficie.

Tip:

* Utiliza la función `round( )` para ajustar los decimales de salida

```python
# Tu código aquí ...
from math import pi

class Esfera():

    def __init__(self, radio):
        self.radio = radio

    def area_superficie(self):
        return round(4 * pi * self.radio ** 2, 2)

    def volumen(self):
        return round(4/3 * pi * self.radio ** 3, 2)

# Solución
```

```python
s = Esfera(1)
```

```python
s.area_superficie()
```

```python
s.volumen()
```

**Este ejercicio lo desarrollaremos en varios pasos.**
1. Crea una clase llamada *GuessingGame*.
2. La clase no necesita ningún argumento proporcionado por el usuario para la instanciación.
3. En el método `.__init__()` (el constructor), haz que la clase establezca un atributo `self.rand_choice` asignándole un número entero aleatorio entre 0-10.
4. Crea un método en la clase llamado *reset_random* que vuelva a generar un número aleatorio (también entre 0-10) para el atributo `self.rand_choice`
5. Crea un método llamado *guess* (adivinar) que use la función `input()` para aceptar un número del usuario. Este método debe imprimir una declaración que indique si la suposición del usuario fue correcta o no.

Bonus track: agrega una lógica en el método *guess* que informe al usuario cuando su número está por arriba o por debajo del número a adivinar.

Tip:
* En el punto 3 use `import random`
* Te dejamos el esqueleto de la clase

```python
# Tu código aquí ...
import random
# Esqueleto de la clase
class GuessingGame():

    def __init__(self):
        self.rand_choice = random.randint(0, 10)

    def reset_random(self):
        self.rand_choice = random.randint(0, 10)

    def guess(self):
        print("Adivina el número entre 0 y 10")
        guess = int(input("Ingrese n a adivinar: "))
        if guess == self.rand_choice:
            print("¡Adivinaste!")
        else:
            if guess > self.rand_choice:
                print("No adivinaste, el número es menor")
            else:
                print("No adivinaste, el número es mayor")

# Solución
```

```python
g = GuessingGame()
```

```python
g.rand_choice
```

```python
g.reset_random()
```

```python
g.guess()
```

```python
g.guess()
```

### Modelando un zoológico
Crea una jerarquía de clases para modelar un zoológico. Debe haber una clase Animal con atributos comunes a todos los animales. Luego, crea clases para diferentes tipos de animales (por ejemplo, Mamífero, Reptil, Ave), cada una de las cuales hereda de Animal. Cada una de estas clases debe tener atributos y métodos específicos de su tipo. Por último, crea clases para animales específicos (por ejemplo, León, Cocodrilo, Águila) que heredan de sus respectivas clases de tipo de animal. Utiliza el polimorfismo para crear un método en la clase Animal que describa el sonido que hace el animal, pero implementa este método de manera diferente en cada clase de animal específico.

```python
# Tu código aquí ...

class Animal():
    def __init__(self, nombre, edad, peso):
        self.nombre = nombre
        self.edad = edad
        self.peso = peso

    def comer(self):
        print(f"{self.nombre} está comiendo.")

    def dormir(self):
        print(f"{self.nombre} está durmiendo.")

    def hacer_sonido(self):
        print(f"{self.nombre} hace un sonido.")  # Método polimórfico, sobrescrito en clases derivadas


# Clase Mamífero, heredada de Animal
class Mamifero(Animal):
    def __init__(self, nombre, edad, peso, pelaje):
        super().__init__(nombre, edad, peso)
        self.pelaje = pelaje  # Específico de Mamífero

    def amamantar(self):
        print(f"{self.nombre} está amamantando a sus crías.")

# Clase Reptil, heredada de Animal
class Reptil(Animal):
    def __init__(self, nombre, edad, peso, temperatura_corporal):
        super().__init__(nombre, edad, peso)
        self.temperatura_corporal = temperatura_corporal  # Específico de Reptil

    def mudar_piel(self):
        print(f"{self.nombre} está mudando su piel.")

# Clase Ave, heredada de Animal
class Ave(Animal):
    def __init__(self, nombre, edad, peso, envergadura_alas):
        super().__init__(nombre, edad, peso)
        self.envergadura_alas = envergadura_alas  # Específico de Ave

    def volar(self):
        print(f"{self.nombre} está volando.")

# Clase León, heredada de Mamífero
class Leon(Mamifero):
    def __init__(self, nombre, edad, peso, pelaje, manada):
        super().__init__(nombre, edad, peso, pelaje)
        self.manada = manada  # Específico de León

    def hacer_sonido(self):
        print(f"{self.nombre} ruge: ¡Roaaar!")

# Clase Cocodrilo, heredada de Reptil
class Cocodrilo(Reptil):
    def __init__(self, nombre, edad, peso, temperatura_corporal, largo):
        super().__init__(nombre, edad, peso, temperatura_corporal)
        self.largo = largo  # Específico de Cocodrilo

    def hacer_sonido(self):
        print(f"{self.nombre} hace un gruñido fuerte.")

# Clase Águila, heredada de Ave
class Aguila(Ave):
    def __init__(self, nombre, edad, peso, envergadura_alas, velocidad):
        super().__init__(nombre, edad, peso, envergadura_alas)
        self.velocidad = velocidad  # Específico de Águila

    def hacer_sonido(self):
        print(f"{self.nombre} emite un chillido agudo.")

# Ejemplos de uso
def main():
    leon = Leon(nombre="Simba", edad=5, peso=190, pelaje="Dorado", manada=True)
    cocodrilo = Cocodrilo(nombre="Snappy", edad=12, peso=500, temperatura_corporal=30, largo=4)
    aguila = Aguila(nombre="Freedom", edad=3, peso=6, envergadura_alas=2.5, velocidad=100)

    # Llamando a métodos específicos de cada animal
    leon.comer()
    leon.hacer_sonido()
    cocodrilo.dormir()
    cocodrilo.hacer_sonido()
    aguila.volar()
    aguila.hacer_sonido()

# Ejecutar el programa
main()
```

### Modelando una biblioteca
Crea una jerarquía de clases para modelar una biblioteca. Debe haber una clase Libro con atributos comunes a todos los libros. Luego, crea clases para diferentes tipos de libros (por ejemplo, Novela, CienciaFiccion, Biografia), cada una de las cuales hereda de Libro. Cada una de estas clases debe tener atributos y métodos específicos de su tipo. Utiliza el polimorfismo para crear un método en la clase Libro que describa el tipo de lectura que proporciona el libro, pero implementa este método de manera diferente en cada clase de libro específico.

```python
# Tu código aqui ...

# Clase base Libro
class Libro:
    def __init__(self, titulo, autor, paginas, publicado):
        self.titulo = titulo
        self.autor = autor
        self.paginas = paginas
        self.publicado = publicado

    def abrir(self):
        print(f"Abriendo el libro '{self.titulo}' de {self.autor}.")

    def cerrar(self):
        print(f"Cerrando el libro '{self.titulo}'.")

    def tipo_de_lectura(self):
        print(f"'{self.titulo}' es un libro de lectura general.")  # Método polimórfico

# Clase Novela, heredada de Libro
class Novela(Libro):
    def __init__(self, titulo, autor, paginas, publicado, genero):
        super().__init__(titulo, autor, paginas, publicado)
        self.genero = genero  # Específico de Novela

    def resumen(self):
        print(f"La novela '{self.titulo}' es del género {self.genero}.")

    def tipo_de_lectura(self):
        print(f"'{self.titulo}' es una novela inmersiva que transporta al lector a otro mundo.")

# Clase CienciaFiccion, heredada de Libro
class CienciaFiccion(Libro):
    def __init__(self, titulo, autor, paginas, publicado, ambientacion):
        super().__init__(titulo, autor, paginas, publicado)
        self.ambientacion = ambientacion  # Específico de CienciaFiccion

    def explorar_conceptos(self):
        print(f"'{self.titulo}' explora conceptos de {self.ambientacion} y ciencia avanzada.")

    def tipo_de_lectura(self):
        print(f"'{self.titulo}' es una obra de ciencia ficción que invita a reflexionar sobre el futuro y la tecnología.")

# Clase Biografia, heredada de Libro
class Biografia(Libro):
    def __init__(self, titulo, autor, paginas, publicado, persona):
        super().__init__(titulo, autor, paginas, publicado)
        self.persona = persona  # Específico de Biografía

    def detalle_de_vida(self):
        print(f"Esta biografía narra la vida de {self.persona}.")

    def tipo_de_lectura(self):
        print(f"'{self.titulo}' es una biografía inspiradora que ofrece una visión profunda de la vida de {self.persona}.")

# Ejemplos de uso
def main():
    novela = Novela(titulo="Orgullo y Prejuicio", autor="Jane Austen", paginas=432, publicado=1813, genero="Romántica")
    ciencia_ficcion = CienciaFiccion(titulo="Dune", autor="Frank Herbert", paginas=688, publicado=1965, ambientacion="un planeta desértico en el futuro")
    biografia = Biografia(titulo="Steve Jobs", autor="Walter Isaacson", paginas=656, publicado=2011, persona="Steve Jobs")

    # Llamando a métodos específicos de cada libro
    novela.abrir()
    novela.resumen()
    novela.tipo_de_lectura()
    ciencia_ficcion.explorar_conceptos()
    ciencia_ficcion.tipo_de_lectura()
    biografia.detalle_de_vida()
    biografia.tipo_de_lectura()

# Ejecutar el programa
main()
```
