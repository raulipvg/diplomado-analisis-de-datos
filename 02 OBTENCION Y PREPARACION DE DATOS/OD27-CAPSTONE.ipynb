````
# Tu código aquí ...
## 1. Reemplace los valores de la columna "Species" por los siguientes valores:
##    "setosa" = 0
##    "versicolor" = 1
##    "virginica" = 2

# Se cambian los valores de la columna "species" segun lo indicado
for i,s in enumerate(iris_obj['target_names']):
  df_iris['Species'] = df_iris['Species'].replace(i,s)

display(df_iris)
````

````
## 2. Agrupe las muestras por especies

## Se agrupa la muestra por especies usando 'Groupby'. Se muestra el promedio
## de los distintos atributos
print(f'Media de atributos de las distintas especies de Iris')
display(df_iris.groupby(by='Species').mean())
print()

# Además, se crean Distintos DataFrames para cada especie, y se elimina la columna
# "Specie" en cada uno por ser información redundante
setosa = df_iris[df_iris['Species']=='setosa'].drop(['Species'], axis=1)
versicolor = df_iris[df_iris['Species']=='versicolor'].drop(['Species'], axis=1)
virginica = df_iris[df_iris['Species']=='virginica'].drop(['Species'], axis=1)

# Se presenta una muestra de los nuevos DataFrames
print('setosa')
display(setosa.sample(2))
print()
print('versicolor')
display(versicolor.sample(2))
print()
print('virginica')
display(virginica.sample(2))
````

````
## 3. Sin usar el método describe calcule los estadísticos básicos de la muestra:
columns = ['sepal length (cm)',	'sepal width (cm)',	'petal length (cm)',	'petal width (cm)']

## cantidad de elementos no nulos
count = df_iris[columns].count(axis=0)
## media
mean = df_iris[columns].mean(axis=0)
## varianza
var = df_iris[columns].var(axis=0)
## Desviación estándar
std= df_iris[columns].std(axis=0)
## Percentiles y cuartiles
P25 = Q1 = df_iris[columns].quantile(0.25, axis=0)
P50 = Q2 = df_iris[columns].quantile(0.5, axis=0)
P75 = Q3 = df_iris[columns].quantile(0.75, axis=0)
## Rango inter-cuartílico
iqr = Q3 - Q1
## Valores máximos y mínimos
max, min = df_iris[columns].max(axis=0), df_iris[columns].min(axis=0)
## Rango (diferencia entre el máximo y el mínimo)
rango = max-min

est_index = ['count','mean','var','std', '25%','50%','75%','IQR','max','min','rango']
estadisticos = [count,mean,var,std,P25,P50,P75,iqr,max,min,rango]
estadistica = pd.DataFrame(estadisticos, index=est_index)
display(estadistica)
````

````
## 4. Repita algunos de los análisis anteriores, pero muestre sus resultados por columnas y especies.

# Se crean y preparan variables para recorrer los datos con un for a traves de agrupaciones segun la especie
columns = ['sepal length (cm)',	'sepal width (cm)',	'petal length (cm)',	'petal width (cm)']
estadisticos_especie = []
especie=[]

## se recorre el for donde se calculan los estadisticos por especie respectiva y
## se guardan en una lispa a partire de la cual se crea un Dataframe, el cual a
## su vez se fuarda en la lista 'estadisticos_especie'
## tambien se guarda el nombre de la especia en la lista 'especie'
for group in df_iris.groupby('Species'):
  count = group[1][columns].count(axis=0)
  mean = group[1][columns].mean(axis=0)
  var = group[1][columns].var(axis=0)
  std = group[1][columns].std(axis=0)
  P25 = Q1 = group[1][columns].quantile(0.25, axis=0)
  P50 = Q2 = group[1][columns].quantile(0.5, axis=0)
  P75 = Q3 = group[1][columns].quantile(0.75, axis=0)
  iqr = Q3 - Q1
  max, min = group[1][columns].max(axis=0), group[1][columns].min(axis=0)
  rango = max-min

  estadisticos = [count,mean,var,std,P25,P50,P75,iqr,max,min,rango]
  estadistica = pd.DataFrame(estadisticos, index=est_index)
  estadisticos_especie.append(estadistica)
  especie.append(group[0])

## se crea un multidex con la especie y los estadisticos para el DataFrame final
est_index = ['count','mean','var','std', '25%','50%','75%','IQR','max','min','rango']
index = pd.MultiIndex.from_product([especie,est_index],names = ['Especie','Estadístico'])

## se crea el dataFrame final a partir de la concatenciacion de los tres
## DataFrames guardados en la lista 'estadisticos_especie' y se le asigna el
## multiindex creado, con lo cual se presentan los estadisiticos separados por
## columna(atributo) y por especie
est_esp = pd.concat(estadisticos_especie)
est_esp.set_index(index, inplace=True)
display(est_esp)
````

````
## 4. Repita algunos de los análisis anteriores, pero muestre sus resultados por columnas y especies.

## Aca se realiza nuevamente la actividad 2.4 pero con una herramienta mucho más
## sencilla de usar (pivot_table.()), aunque la visualización de los estadisticos
## de esta manera no es tan cómoda. Se debió generar 2 tablas pivot distintas
## para poder ver todos los estadisticos
display(df_iris.pivot_table(index = "Species", aggfunc = ['count','mean', 'var', 'std',]))
print()
display(df_iris.pivot_table(index = "Species", aggfunc = ['quantile','max','min']))
````

````
## 5. Compare sus resultados usando el metodo describe por columnas y por especies
df_iris.groupby('Species').describe()
````

Comparando los resultados con el método describe por columnas y especies, notamos que la manera de presentar la información es muy similar a la de `pivot_table`, es decir, hacia el lado, con lo que la información no es tan accesible ni fácil de ver. Sin embargo, el orden de las columnas es diferente, dado que usando `pivot_table` se tiene en el nivel 0 de las columnas el estadístico y en el nivel 1 el atributo. Al contrario, utilizando la función `groupby` junto a la función `describe`, donde el nivel 0 de las columnas presenta el atributo y, por su parte, el nivel 1 indica el estadístico.

5. Para una mejor interpretación de la información anterior, genere una tabla con el número de entradas por cada grupo.






````
# Tu código aquí ...
# Dividir la columna 'age' en intervalos especificados
def tramo_edad(age):
    if age <= 12:
        return '0-12'
    elif 12 < age <= 18:
        return '12-18'
    elif 18 < age <= 80:
        return '18-80'
    else:
        return 'Unknown'  # Para edades fuera del rango especificado o NaN
titanic['tramo_etario'] = titanic['age'].apply(tramo_edad)

# Verificar los resultados
titanic.sample(10)
````

````
# Agrupar por 'sex', 'pclass', y 'age_group'
grouped_data = titanic.groupby(['sex', 'pclass', 'tramo_etario'])['survived'].agg(['mean', 'sum'])

# Renombrar las columnas para mayor claridad
grouped_data.rename(columns={'mean': 'Tasa de supervivencia promedio', 'sum': 'Cantidad de sobrevivientes'}, inplace=True)

display(grouped_data)
````

````
# Tu código aquí ...
# Calcular el valor mediano de la columna 'fare'
boleto_mediana = titanic['fare'].median()

# Categorizar los precios de los boletos basados en el valor mediano
def categoria_boleto(fare, boleto_med):
    if fare > boleto_med:
        return 'Caros'
    else:
        return 'Baratos'

# Categorizar los precios de los boletos utilizando la función definida
titanic['categoria_boleto'] = titanic['fare'].apply(categoria_boleto, args=(boleto_mediana,))

# Verificar los resultados
print(f"La mediana de los boletos es: {boleto_mediana}\n")
titanic.sample(10)
````

````
# Crear una tabla pivote para genero, clase, precios
pivot_precios = titanic.pivot_table(
    values='survived',
    index=['sex', 'pclass'],
    columns='categoria_boleto',
    aggfunc=('mean','sum','size')
)
pivot_precios.fillna('-', inplace=True)
display(pivot_precios)
````

Para cada clase y sexo hay un mayor número de boletos baratos comprados que caros, pero tambien hay un mayor número de sobrevivientes mujeres que hombres para cada clase y precio.
