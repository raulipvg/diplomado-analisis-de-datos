# <b>Algunos conceptos de BD Relacionales.</b>
<ul>
<li> Una base de datos es conjunto de archivos lógicamente relacionados. </li>
<li> Los archivo se representan en forma de tablas bidimensionales (filas y columnas).</li>
<li><b>Fila:</b> excepto la primera fila, se conoce como registro, tupla, fila.
<ul>
<li>Una fila esta compuesta por valores de distinto tipo.</li>
<li>El orden de las filas es irrelevante.</li>
<li>Cada fila tiene un valor que permite identificarla de manera única.</li>
</ul>
</li>
<li><b>Columna:</b> se conoce como campo, atributo, columna.
<ul>
<li>En una tabla, el nombre de cada columna es único.</li>
<li>Todos los valores de una columna son del mismo tipo.</li>
<li>El orden de las columnas es irrelevante.</li>
</ul>
</li>
<li><b>Importante:</b> Se dice que los datos son estructurados.</li>
</ul>

#<b>La sentencia SELECT</b>
* Para recuperar datos (consultas) desde una o más tablas usaremos la sentencia [SELECT](https://sqlite.org/lang_select.html). La sentencia SELECT se compone de 6 cláusulas: <b>`SELECT`</b>, <b>`FROM`</b>, <b>`WHERE`</b>, <b>`GROUP BY`</b>, <b>`HAVING`</b> y <b>`ORDER BY`</b>.
 * `SELECT` especifica las columnas que se incluirán en el resultado de la consulta.
 * `FROM` especifica las tablas desde las cuales se recuperarán los datos y como se realizarán las uniones (`JOIN`) de tablas.
 * `WHERE` se utiliza para filtrar filas desde las tablas.
 * `GROUP BY` se utiliz para agrupar filas en baase a valores comunes.
 * `HAVING` se utiliza para filtrar grupos creador con `GROUP BY`.
 * `ORDER BY` ordena las filas de resultado final en base a los valores de una o más columnas incluidas en él.

##<b>SELECT</b>

* __`SELECT * FROM Persona`__, recupera todas las columnas y todas las filas desde la tabla _test_.
* __`SELECT Rut, Nombre FROM Persona`__, recupera todas las filas desde la tabla *tes_t* pero incluye solamente las columnas _id_ y _nombre_.
* __`SELECT UPPER(Nombre) FROM Persona`__, recupera todas filas desde la tabla **Persona** y al mismo tiempo los valores de la columna **Nombre** los muestra en mayúsculas.  


````
cursor.execute("""
                SELECT *
                FROM Persona
              """)
print(cursor.fetchall())
````

````
cursor.execute("""
                SELECT Rut,Nombre
                FROM Persona
              """)
print(cursor.fetchall())
````

````
cursor.execute("""
                SELECT UPPER(Nombre)
                FROM Persona
              """)
print(cursor.fetchall())
````

SQLite tiene un catálogo que contiene los nombres de las tablas que hemos creado en nuestra base de datos. Este catálogo se encuentra en una tabla llamada `sqlite_master`.

````
cursor.execute("""
                SELECT Name
                FROM sqlite_master
                WHERE type='table';
              """) #Seleccionamos la columna Name de la tabla sqlite_master, donde la columna type sea igual a 'table'
print(cursor.fetchall()) #Le pedimos a nuestro cursor que obtenga todos los resultados y los muestre
````

Para saber cual es la estructura de una tabla se utiliza la sentencia `PRAGMA`

````
cursor.execute("PRAGMA table_info('Persona')")
for el in cursor.fetchall():
    print(el)
````

##<b> WHERE</b>

La cláusula __WHERE__ nos permite ser selectivos y aplicar filtros

* __`SELECT * FROM Persona WHERE Edad < 20`__ , recupera todas las columnas de todas las filas cuyo valor en la columna edad sea menor que 20.
* __`SELECT * FROM Persona WHERE Nombre = 'Pepe Lota'`__, recupera todas las columnas de todas las filas cuyo valor en la columna nombre sea igual a 'Pepe Lota'.


````
cursor.execute("""
                SELECT *
                FROM Persona
                WHERE Edad < 40
              """)
print(cursor.fetchall())
````

````
cursor.execute("""
                SELECT *
                FROM Persona
                WHERE Nombre='Harrison Spencer'
              """)
print(cursor.fetchall())
````

````
cursor.execute("""
                SELECT *
                FROM Persona
                WHERE Nombre like '%Bull%'
              """)
print(cursor.fetchall())
````

Podemos borrar/eliminar tablas usando DROP:

````
cursor.execute("DROP table Persona")
````

##**Consulta 3**

````
#Su codigo aquí
query = """
          SELECT \"Tipus_Vehicles\", \"Nom_barri\", \"Nom_districte\"
          FROM padron
          WHERE Tipus_Vehicles = "Furgonetes"
        """
pd.read_sql(query,connDF)
````

##**Consulta 4**

Seleccionamos Tipos de Vehiculos filtrando por su sello ambiental: "Etiqueta C – Verda"

````
#Su codigo aquí
query = """
          SELECT  DISTINCT \"Tipus_Vehicles\"
          FROM padron
          WHERE "Desc_distintiu_ambiental" = "Etiqueta C – Verda"
        """
tipos_vehiculos = pd.read_sql(query,connDF)['Tipus_Vehicles'].to_list()

pares = zip(tipos_vehiculos,['Turismo', 'Motocicletas','Ciclomotores', 'Furgonetas','Camiones', 'Otros vehículos'])
traductor = dict(pares)
````
