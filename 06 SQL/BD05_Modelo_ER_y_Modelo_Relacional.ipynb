# <b> El Modelo Entidad Relación (MER) </b>
<ul>
<li>Es una representación de tablas-datos-relaciones que se almacenarán en una BD:</li>
<li>Es independiente del hardware y del software.</li>
<li> La notación que se utiliza puede ser comprendida tanto por el usuario/cliente como por el diseñador de la BD.</li>
<li>El modelo ER fue propuesto por Peter Chen en su artículo <a href="https://dspace.mit.edu/bitstream/handle/1721.1/47432/entityrelationshx00chen.pdf"> The Entity-Relationship Model-Toward a Unified View of Data</a>, ACM Transactions on Database Systems, Vol 1, Nro 1, March 1976.</li>
<li>Se basa en dos conceptos: Entidad y Relación</li>
</ul>

#**Entidad**
<ul>
<li>Es todo sobre lo cual se desea recolectar datos y almacenarlos.</li>
<li>En general son conceptos reconocibles, concretos o abstractos tales como: persona, lugar, proyecto, etc.</li>
<li>Una ocurrencia individual de una entidad es un instancia. Por ejemplo: existe una entidad <b>Alumno</b> y miles instancias de esta entidad.</li>
<li>Se recomienda que el nombre de una entidad se escriba en singular.</li>
<li>Ejemplos de entidades son los siguientes:
  
      Persona: Estudiante, Empleado, Cliente, Paciente.
      Lugar: Ciudad, Habitación.
      Evento: Concierto, Exposición, Guerra.
      Objeto: Máquina, Auto, Barco.
      Concepto: Proyecto, Asignatura, Cuenta.
</li>

#**Atributos**

<ul>
<li>Una entidad es descrita mediante sus atributos. Algunos ejemplos de atributos son los sigientes:

      Empleado = {Rut, Nombre, Apellido, Edad}
      Ciudad = {Nombre, Población, Región}
      Concierto = {Fecha, Hora, Lugar}
      Proyecto = {IdProyecto, Nombre, Fecha Inicio, Fecha Cierre}
</li>
<li> Cuando todos los atributos tienen valores entonces originan una instancia
de una entidad. Por ejemplo:

      Empleado = {20.227.888-0, Mike Patton, 52}
</li>

<li> Existen diferentes tipos de atributos:
<ul>

*   **Simple** (o atómico): no se puede dividir en otros componentes. Ejemplo: Rut, Población, IdProyecto.
*   **Compuesto**: se puede dividir y cada una de sus partes es un atributo simple. Ejemplo: Name y Address (ver figura de más abajo).
*   **Univaluados**: posee un único valor. Ejemplo: Ciudad, Rut, Nombre, Apellido.    
*   **Multivaluados**: son atributos que podrían tener uno o más valores. En la figura de más abajo el atributo Language es multivaluado ya que almacena los lenguajes que habla un estudiante.
*   **Derivados**:un atributo derivado obtiene su valor a partir de otro(s) atributo(s). Ejemplo: Age se puede calcular a partir de DateofBirth.
*   **Clave** (o identificador): permite identificar de manera única a una instancia de una entidad. Ejemplo: Rut, Patente, Rol (de una propiedad), Código (de un producto), Patente (vehículo, o giro comercial).
</ul>
</li>
</ul>

#**Cardinalidad**
La cardinalidad refiere al número de instancias de una entidad que pueden
estar relacionadas con las instancias de otra. No es necesario un número
preciso, la máxima cardinalidad se expresa en términos generales por
medio de cuatro tipos de relaciones: Uno-a-Uno, Uno-a-Muchos,
Muchos-a-Uno y Muchos-a-Muchos.

##**Relaciones uno-a-uno (1:1)**.
Una instancia de la entidad A se relaciona sólo con una instancia de la entidad B. Por ejemplo: Un país tiene una capital y una capital pertenece a un país.

<center><img src="https://drive.google.com/uc?id=118B877_HjgZkkTQJ4gP0M57TXnxa5Lsa" alt="drawing" width=600px/></center>

##**Relaciones uno-a-muchos (1:N).**
 Una instancia de la entidad A se relaciona con muchas instancias de la entidad B, al mismo tiempo una instancia de B se relaciona con una instancia de A. Por ejemplo: un cliente realiza muchas compras pero una compra es realizada sólo por un cliente.

 <center><img src="https://drive.google.com/uc?id=17VVC6s-VSCq4KZJSxjfuHrCDyUHKxEzx" alt="drawing" width=600px/></center>

##**Relaciones de muchos-a-muchos (N:M).**
Una instancia de la entidad A se relaciona con muchas instancias de la entidad B y una instancia de la entidad B se relaciona con muchas instancias de la entidad A. Por ejemplo: un ingeniero trabaja en muchos proyectos y en un proyecto trabajan muchos ingenieros.

<center><img src="https://drive.google.com/uc?id=1PFazpxUKaRum8E4yR0P7NpJFiisRGrc1" alt="drawing" width=600px/></center>


<ul>
<li>Una relación N:M no se puede implementar directamente en el modelo relacional.</li>

<li>Se transforma en 2 relaciones del tipo 1:N y se genera una entidad de asociación.</li>

<li>Más allá de una cuestión de implementación, en algunos casos es necesario almacenar datos que pertenecen a la relación. por ejemplo: Los empleados pueden ser asignados a muchos proyectos y en un proyecto trabajan muchos empleados. Si un requerimiento fuera "calcular la cantidad de horas trabajados por proyecto de un determinado empleado". Donde registramos las horas trabajadas?</li>
</ul>

#**Del MER al Modelo Relacional**

*   Una vez finalizado, el MER se puede transformar al Modelo Relacional (tablas, filas, columnas, claves, etc).
*   En general cada entidad del MER se transforma en una tabla del modelo relacional, y cada uno de los atributos de una entidad se transforman en los nombres de las columnas (atributos) de las tablas.
*   A partir del atributo clave de una entidad y la cardinalidad de una relación entre entidades, surgen los dos atributos fundamentales del modelo relacional: la clave primaria y la clave foránea.   



#**Clave primaria y clave foránea**
<ul>
<li>Corresponden al componente más básico sobre el cual se basa el modelo relacional.</li>
<li>La clave primaria identifica de manera única las ocurrencias de una entidad.</li>
<ul>
<li>Toda entidad debe tener una clave primaria.</li>
<li>Puede estar compuesta de uno o más atributos.</li>
<li>Su valor es único para cada instancia y no permite valores nulos.</li>
<li>Una clave primaria puede ser "artificial".</li>
</ul>

<li>La clave foránea permite materializar una asociación entre 2 entidades.</li>
<ul>
<li>En una relación 1:N entre las entidades A y B, la entidad B "hereda" la clave primaria de la entidad A.</li>
<li>En la entidad B, la clave "heredada" se conoce como clave foránea.</li>
<li>La clave foránea permite "navegar" en el modelo relacional.</li>
</ul>

</ul>

##**Proponga 3 consultas más:**

1. ¿Cuántos estudiantes se inscribieron en cada curso durante el año 2021?

````
# Consulta 1

query="""
        SELECT c.curso_id, c.nombre AS curso_nombre, COUNT(i.estudiante_id) AS cantidad_inscritos
        FROM Inscripcion i
        JOIN Curso c ON i.curso_id = c.curso_id
        WHERE i.anio = 2021
        GROUP BY c.curso_id, c.nombre;
      """

pd.read_sql(query,conn)
````

2. ¿Cuál es el promedio de notas obtenidas en cada curso durante el año 2020?


````
# Consulta 2
query="""
        SELECT c.curso_id, c.nombre AS curso_nombre, AVG(i.nota) AS promedio_notas
        FROM Inscripcion i
        JOIN Curso c ON i.curso_id = c.curso_id
        WHERE i.anio = 2020 AND i.nota IS NOT NULL
        GROUP BY c.curso_id, c.nombre;
        """

pd.read_sql(query,conn)
````

## Identificar Estudiantes con Múltiples Inscripciones en el Mismo Semestre (Año 2020)

Este experimento analiza la carga académica de los estudiantes en el año 2020, identificando aquellos que se inscribieron en más de un curso dentro del mismo semestre. Utilizando GROUP BY y HAVING, la consulta agrupa las inscripciones por estudiante y semestre, contando la cantidad de asignaturas inscritas.

````
query="""
        SELECT
            e.estudiante_id, e.nombres,e.apellidos,i.semestre, COUNT(i.curso_id) AS total_cursos_inscritos
        FROM Inscripcion i
        JOIN Estudiante e ON i.estudiante_id = e.estudiante_id
        WHERE i.anio = 2020
        GROUP BY e.estudiante_id, i.semestre
        HAVING total_cursos_inscritos > 1;
      """

pd.read_sql(query,conn)
````

## Gestión de Inscripciones Académicas: Creación y Reemplazo de Vistas en SQLite

Este experimento demuestra cómo crear y reemplazar una vista en SQLite para gestionar eficientemente los datos de inscripciones académicas. Se utiliza DROP VIEW IF EXISTS para eliminar la vista si ya existe y luego se recrea con información consolidada de los estudiantes, cursos y sus inscripciones en el año 2020. Esto facilita consultas recurrentes sin necesidad de repetir uniones complejas en cada consulta.

````
cursor.execute("DROP VIEW IF EXISTS vista_inscripciones_2020;")
query="""
        CREATE  VIEW vista_inscripciones_2020 AS
      SELECT
          e.estudiante_id,
          e.nombres,
          e.apellidos,
          c.curso_id,
          c.nombre AS curso_nombre,
          c.clave,
          i.semestre,
          i.anio,
          i.nota
      FROM Inscripcion i
      JOIN Estudiante e ON i.estudiante_id = e.estudiante_id
      JOIN Curso c ON i.curso_id = c.curso_id
      WHERE i.anio = 2020;
      """
cursor.execute(query)
conn.commit()
##pd.read_sql(query,conn)

````

Este experimento demuestra cómo consolidar datos de múltiples tablas en una vista, facilitando el acceso y el análisis de la información para un periodo específico.
