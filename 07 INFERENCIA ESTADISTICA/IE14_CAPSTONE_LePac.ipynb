```python
# Tu c√≥digo aqu√≠ ...
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Tu c√≥digo aqu√≠ ...

# X = Satisfacci√≥n ultimo trimestre de Capel, sigue una Distribuci√≥n Binomial
# Definir par√°metros ùëã~Bin(n,p)
mu_capel = 4.5 # mu = n*p
i_capel = 7 # "intentos/tama√±o muestra"
p_capel = mu_capel/i_capel # probabilidad de √©xito, mu = n*p

binomial_capel = stats.binom(i_capel, p_capel) # Distribuci√≥n
x_capel = np.arange(binomial_capel.ppf(0.01),
              binomial_capel.ppf(0.99)+1)
fmp_capel = binomial_capel.pmf(x_capel) # Funci√≥n masa de probabilidad
respuestas_capel = np.random.binomial(i_capel, p_capel, 800)
n_capel = len(respuestas_capel)
print(f"Cantidad de respuestas a encuenta satisfacci√≥n Capel, √∫ltimo trimestre : {n_capel}\n")
print(respuestas_capel)
```

```python
# X = Satisfacci√≥n ultimo trimestre de competencia, sigue una Distribuci√≥n Binomial
# Definir par√°metros ùëã~Bin(n,p)
mu_competencia = 6.1 # mu = n*p
i_competencia = 7 # "intentos/tama√±o muestra"
p_competencia = mu_competencia/i_competencia # probabilidad de √©xito, mu = n*p

binomial_competencia = stats.binom(i_competencia, p_competencia) # Distribuci√≥n
x_competencia = np.arange(binomial_competencia.ppf(0.01)-3,
              binomial_competencia.ppf(0.99)+1)
fmp_competencia = binomial_competencia.pmf(x_competencia) # Funci√≥n masa de probabilidad
respuestas_competencia = np.random.binomial(i_competencia, p_competencia, 500)
n_competencia = len(respuestas_competencia)
print(f"Cantidad de respuestas a encuenta satisfacci√≥n Competencia, √∫ltimo trimestre : {n_competencia}\n")
print(respuestas_competencia)
```

```python
# X = Satisfacci√≥n ultimo trimestre de competencia, sigue una Distribuci√≥n Binomial =>
# Definir par√°metros ùëã~Bin(n,p)
mu_lepac = 6.15 # mu = n*p
i_lepac = 7 # "intentos/tama√±o muestra"
p_lepac = mu_lepac/i_lepac # probabilidad de √©xito, mu = n*p

binomial_lepac = stats.binom(i_lepac, p_lepac) # Distribuci√≥n
x_lepac = np.arange(binomial_lepac.ppf(0.01)-3,
              binomial_lepac.ppf(0.99)+1)
fmp_lepac = binomial_lepac.pmf(x_lepac) # Funci√≥n masa de probabilidad
respuestas_lepac = np.random.binomial(i_lepac, p_lepac, 320)
n_lepac = len(respuestas_lepac)
print(f"Cantidad de respuestas a encuenta satisfacci√≥n Le Pac, √∫ltimo trimestre : {n_lepac}\n")
print(respuestas_lepac)
```

```python
# 3. Calcular top-2-box-score
counter_capel = 0
counter_competencia = 0
counter_lepac = 0

for i in range(len(respuestas_capel)):
  if respuestas_capel[i] == 6 or respuestas_capel[i] == 7:
    counter_capel += 1

for i in range(len(respuestas_competencia)):
  if respuestas_competencia[i] == 6 or respuestas_competencia[i] == 7:
    counter_competencia += 1

for i in range(len(respuestas_lepac)):
  if respuestas_lepac[i] == 6 or respuestas_lepac[i] == 7:
    counter_lepac += 1

top2box_capel = counter_capel/len(respuestas_capel)
top2box_competencia = counter_competencia/len(respuestas_competencia)
top2box_lepac = counter_lepac/len(respuestas_lepac)

print(f"Top2Box Capel:       {top2box_capel:4.2%}")
print(f"Top2Box Competencia: {top2box_competencia:4.2%}")
print(f"Top2Box Le Pac:      {top2box_lepac:4.2%}")
```

```python
# 4. Calcular intervalo de confianza para la media y proporci√≥n Top2Box con significancia 0.05.

alpha = 0.05
NC = 1 - alpha
alpha_medio = alpha / 2

# Calcular Z con la funci√≥n de distribuci√≥n normal acumulada inversa (ppf)
Z = stats.norm.ppf(1 - alpha_medio)
print(f"Z: {Z: 6.4f}")

# Calcular intervalos de confianza media
# como n >= 30 aplicaremos Z (Distribucion normal estandarizada)
LI_xbar_capel = xbar_capel - Z * (S_capel / np.sqrt(n_capel))
LS_xbar_capel = xbar_capel + Z * (S_capel / np.sqrt(n_capel))

LI_xbar_competencia = xbar_competencia - Z * (S_competencia / np.sqrt(n_competencia))
LS_xbar_competencia = xbar_competencia + Z * (S_competencia / np.sqrt(n_competencia))

LI_xbar_lepac = xbar_lepac - Z * (S_lepac / np.sqrt(n_lepac))
LS_xbar_lepac = xbar_lepac + Z * (S_lepac / np.sqrt(n_lepac))

# Calcular intervalos de confianza  Top2Box

# Se cumple n>30 para Capel, competencia y Le Pac.
# Ahora verifiquemos que se cumple n*p>5 y n*(1-p)>5
print()
print(f"Capel n*p:            {n_capel * top2box_capel:4.2f}")
print(f"Capel n*(1-p):        {n_capel * (1 - top2box_capel):4.2f}")
print()
print(f"Competencia n*p:      {n_competencia *top2box_competencia:4.2f}")
print(f"Competencia n*(1-p):  {n_competencia * (1 - top2box_competencia):4.2f}")
print()
print(f"Le Pac n*p:           {n_lepac *top2box_lepac:4.2f}")
print(f"Le Pac n*(1-p):       {n_lepac * (1 - top2box_lepac):4.2f}")
print()
#Se cumple que n*p>5 y n*(1-p)>5 para Capel, competencia y Le Pac.


# como n >= 30 aplicaremos Z (Distribucion normal estandarizada)
LI_top2box_capel = top2box_capel - Z * (np.sqrt(top2box_capel * (1 - top2box_capel)) / np.sqrt(n_capel))
LS_top2box_capel = top2box_capel + Z * (np.sqrt(top2box_capel * (1 - top2box_capel)) / np.sqrt(n_capel))

LI_top2box_competencia = top2box_competencia - Z * (np.sqrt(top2box_competencia * (1 - top2box_competencia)) / np.sqrt(n_competencia))
LS_top2box_competencia = top2box_competencia + Z * (np.sqrt(top2box_competencia * (1 - top2box_competencia)) / np.sqrt(n_competencia))

LI_top2box_lepac = top2box_lepac - Z * (np.sqrt(top2box_lepac * (1 - top2box_lepac)) / np.sqrt(n_lepac))
LS_top2box_lepac = top2box_lepac + Z * (np.sqrt(top2box_lepac * (1 - top2box_lepac)) / np.sqrt(n_lepac))
print()
print(f'Intervalo de Confianza (IC) al {NC} para media poblacional Capel:       {LI_xbar_capel:4.2f}, {LS_xbar_capel:4.2f}')
print(f'Intervalo de Confianza (IC) al {NC} para media poblacional Competencia: {LI_xbar_competencia:4.2f}, {LS_xbar_competencia:4.2f}')
print(f'Intervalo de Confianza (IC) al {NC} para media poblacional Le Pac:      {LI_xbar_lepac:4.2f}, {LS_xbar_lepac:4.2f}')
print()
print(f'Intervalo de Confianza (IC) al {NC} para Top2Box Capel:       {LI_top2box_capel:4.2%}, {LS_top2box_capel:4.2%}')
print(f'Intervalo de Confianza (IC) al {NC} para Top2Box Competencia: {LI_top2box_competencia:4.2%}, {LS_top2box_competencia:4.2%}')
print(f'Intervalo de Confianza (IC) al {NC} para Top2Box Le Pac:      {LI_top2box_lepac:4.2%}, {LS_top2box_lepac:4.2%}')
```

## <font color='green'>**Actividad 3**</font>
### Creando nuestro test de hip√≥tesis

Desarrollar un test de hip√≥tesis de 1 muestra para al menos 1 de los 3 datasets. Aqu√≠ Ud. puede formular la hip√≥tesis a probar. Por ejemplo, puede ser contra un target definido por la industria o retail.



**Hip√≥tesis:** Se desea contrastar con un nivel de significaci√≥n del 5 % la hip√≥tesis de que la **satisfacci√≥n media del grupo Capel es mayor o igual a 5.0**, considerado por la empresa como el KPI para recibir el bono anual de desempe√±o.


```python
# Definir hip√≥tesis a contrastar (prueba de cola izquierda)
#    H0:   Mu >= 5.0
#    H1:   Mu < 5.0

# n > 30, usaremos estad√≠stico de prueba Z

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Par√°metros del problema
mu_0 = 5          # Media bajo la hip√≥tesis nula
x_bar = np.mean(respuestas_capel)       # Media muestral observada
sigma = np.std(respuestas_capel, ddof=1) # Desviaci√≥n est√°ndar muestral insesgada
n = np.size(respuestas_capel)          # Tama√±o muestra

alpha = 0.05      # Nivel de significancia

# C√°lculos
se = sigma / np.sqrt(n)                    # Error est√°ndar
Z = (x_bar - mu_0) / se                    # Estad√≠stico Z
Z_crit = norm.ppf(alpha)               # Valor cr√≠tico Z (Z de tabla)
p_value = 1 - norm.cdf(Z)                  # Valor p

# Datos para la distribuci√≥n de la media muestral
x_dist = np.linspace(mu_0 - 1, mu_0 + 1, 1000) #se crea un rango de valors para el eje x desde 1 a 15, con 1000 valores para suavizar la curva
y_dist = norm.pdf(x_dist, mu_0, se) #se grafica la distribuci√≥n de las medias muestrales centrada en mu = 5.5

# Datos para la distribuci√≥n normal est√°ndar
x_std = np.linspace(-2, 2, 1000)
y_std = norm.pdf(x_std, 0, 1)

# Crear subplots verticales
fig, axs = plt.subplots(figsize=(10, 5))

# -------- GR√ÅFICO 1: Distribuci√≥n de la media muestral --------
axs.plot(x_dist, y_dist, color='blue', label='Distribuci√≥n bajo $H_0$: $\mu >= 5.0$')
axs.axvline(mu_0, color='black', linestyle=':', label='Media bajo $H_0$')
axs.axvline(x_bar, color='green', linestyle='--', label=f'Muestra: $\overline{{x}} = {x_bar}$')
axs.axvline(mu_0 + Z_crit * se, color='red', linestyle='--', label=f'Z cr√≠tico = {Z_crit:.2f}')
axs.fill_between(x_dist, 0, y_dist, where=(x_dist >= mu_0 + Z_crit * se), color='red', alpha=0.2, label='Zona de rechazo')
axs.fill_between(x_dist, 0, y_dist, where=(x_dist >= x_bar), color='purple', alpha=0.3, label=f'Valor p ‚âà {p_value:.4f}')
axs.set_title('Distribuci√≥n de la media muestral (en puntos)')
axs.set_xlabel('Media muestral (puntos)')
axs.set_ylabel('Densidad de probabilidad')
axs.legend()
axs.grid(True)
```

**Por lo tanto, se rechaza H0, con lo cual la media es menor que 5.0, no cumpliendo el KPI para ese a√±o.**

```python
# Tu c√≥digo aqu√≠ ...

##########################################################################
##Verificar si la satisfacci√≥n de Le Pac es mayor a la del producto Capel.
##########################################################################

# An√°lisis post hoc utilizando t Student, con varianza desigual

## 1) cargar datos muestras: tama√±o (n), media (xbarra) y sigma (S)
# muestra 1 (experimental)
n_1 = n_lepac
xbarra_1 = xbar_lepac
S_1 =  S_lepac

# muestra 2 (control)
n_2 = n_capel
xbarra_2 = xbar_capel
S_2 = S_capel

## 2) Determinar Nivel de Significancia
alpha = 0.05

## 3) Estimar el IC de la diferencia de medias

# Estimar grados de libertad cuando varianzas poblacionales no son iguales
gl = (S_1**2 /n_1 + S_2**2 /n_2)**2  /  (((S_1**2/n_1)**2 / (n_1-1)) + ((S_2**2/n_2)**2 / (n_2-1)))
print(f'Estimaci√≥n gl: {gl:4.2f}')

# calcular T critico 2 colas con Scipy
Tcritico = stats.t.ppf(1 - (alpha/2), df=gl)
print (f'T critico: {Tcritico:5.3f}')


# Estimar ponderado de la varianza poblacional (S pooled) =>  #NO APLICA para calcular IC ya que Varianzas son diferentes
# Se aplica la varianza combinada para calcular 'd Cohen' en el √∫ltimo paso, por esta raz√≥n se calcula aqu√≠:
var = ( S_1**2 *(n_1 - 1) + S_2**2 *(n_2 - 1)) / (n_1 + n_2 - 2)
S = np.sqrt(var)
print(f'Varianza combinada: {var:5.3f}')
print(f'Sigma poblaci√≥n: {S:5.3f}')


# como las varianzas son diferentes, el c√°lculo del 'Sigma diferencias  medias' es de la siguiente forma:
S_dif = np.sqrt( (S_1**2 / n_1) + (S_2**2 / n_2) )
print(f'S diferencia medias: {S_dif:5.3f}')

# IC diferencia medias al 1-alpha:
LI = (xbarra_1 - xbarra_2) - Tcritico * S_dif
LS = (xbarra_1 - xbarra_2) + Tcritico * S_dif
print(f'IC diferencia medias al {1-alpha:5.3f} NC: {LI:5.3f}    {LS:5.3f}')

# otra forma con Scipy
IC = stats.t.interval(1-alpha, loc=xbarra_1 - xbarra_2, scale=S_dif, df=gl)
print(f'IC diferencia medias al {1-alpha:5.3f} NC: {IC[0]:5.3f}   {IC[1]:5.3f}')
```

```python
# Otra forma de estimar grados de libertad con varianzas desiguales es con la la ecuaci√≥n Welch‚ÄìSatterthwaite.
# N√≥tese que la verdadera distribuci√≥n de este estad√≠stico de hecho depende (ligeramente) de dos varianzas desconocidas.
gll = (S_1**2 /n_1 + S_2**2 /n_2)**2  /  (((S_1**2/n_1)**2 / (n_1-1)) + ((S_2**2/n_2)**2 / (n_2-1))) - 2
print(f'Grados de Libertad: {gll:5.3f}')
```

```python
## 4) plantear hip√≥tesis
 # H0:  Mu 1 = Mu 2
 # H1:  Mu 1 ‚â† Mu 2

# regla decisi√≥n (2 colas)
 # no rechazar H0 si: -Tcritico <= T <= Tcritico
 # rechazar HO si: T < -Tcritico o T > Tcritico

## 5) Calcular el estad√≠stico T cr√≠tico a emplear
Tcritico = Tcritico #fue obtenido en paso anterior

## 6) Determinar el tipo de test, una o dos colas.
# como la hip√≤tesis alternativa es del tipo ‚Äúdiferente de‚Äù se trata de un test de dos colas
# Nota: Se recomienda utilizar el contraste bilateral para estos an√°lisis dado que √©stos
# requieren una mayor diferencia para detectar una misma potencia

## 7 ) Calcular el estad√≠stico T de prueba
# como varianzas no son iguales, aqui el denominador es el sigma de diferencias calculado en paso #3
T = ( xbarra_1 - xbarra_2 - 0) / S_dif
print(f'T prueba: {T:5.3f}')

## 8) Calcular el p-value y comparaci√≥n con el nivel de significancia establecido.
# calcular p-value
pvalue = 1 - (stats.t.cdf(T, df=(gl))) # T prueba se encuentra en cola derecha
print(f'p-value (significancia estad√≠stica): {pvalue*2:7.5f}') # 2 colas multiplicar por 2

## 9) Concluir prueba de hip√≥tesis:
# como T prueba > Tcritico o como p-value menor a alpha, se rechaza H0, las medias son diferentes.
```

```python
## 10) Calcular la potencia de la prueba (1 - beta)

# buscar el valor en d√≠as correspondiente al T critico
x = 0 + (Tcritico * S_dif)# por la derecha
# print('X promedio para T=',Tcritico,': ',x)

# otra forma mas simple con Scipy usando ppf
x_b = stats.t.ppf((1-alpha/2), df=gl , loc=0, scale=S_dif)
# print('X promedio para T=',Tcritico,': ',x_b)


# Calcular Beta (error tipo II)
# forma simple con Scipy
beta = stats.t.cdf(x, df=gl, loc=xbarra_1-xbarra_2, scale=S_dif)
print(f'Beta: {beta:5.3f}')

# otra forma simple para calcular Beta con Scipy
beta1 = stats.t.cdf(Tcritico-T, df= gl)
print(f'Otra forma beta: {beta1:5.3f}')

# Calcular Potencia estad√≠stica (1- beta)
print(f'Potencia de la prueba: {1-beta:5.3f}')


## 11) Calcular el tama√±o del efecto
# aqu√≠ utilizaremos la desv. combinada (S pooled)
# d de Cohen para T-TEST INDEPENDIENTES: (diferencia medias grupos / S medias 2 sigmas)

dif_bruta = (xbarra_1 - xbarra_2)
dCohen = (xbarra_1 - xbarra_2) / S
print(f'd Cohen: {dCohen:5.3f}')
print(f'Diferencia bruta: {dif_bruta:5.3f}')
```

```python
# graficar
#  Graficando la distribuci√≥n de las diferencias de medias
n = n_1 + n_2
xbar = xbarra_1 - xbarra_2
S_dif = S_dif

# proyectar distribuci√≥n diferencias medias (H1)
dif_medias_H1 = stats.t(loc=xbar, scale=S_dif, df=(n-1))
x = np.linspace(dif_medias_H1.ppf(0.0001), dif_medias_H1.ppf(0.9999), 100)
pdf = dif_medias_H1.pdf(x) # funci√≥n densidad probabilidad

# proyectar distribuci√≥n H0
dif_medias_H0 = stats.t(loc=0, scale=S_dif, df=(n-1))
x_m = np.linspace(dif_medias_H0.ppf(0.0001), dif_medias_H0.ppf(0.9999), 100)
pdf_m = dif_medias_H0.pdf(x_m) #f unci√≥n densidad probabilidad

plt.figure(figsize=(10, 5))
plt.ylim(0, pdf.max()+3)

plt.yticks([])

plt.plot(x_m, pdf_m,"b",label=r'$H_0$' ) # dif medias H0 fit
plt.plot(x, pdf,"orange", label=r'$H_1$') # dif medias H1 fit
plt.vlines(ymin=0, ymax=pdf.max(), x=-Tcritico*S_dif,
           color='r') # LI IC
plt.vlines(ymin=0, ymax=pdf.max(), x=Tcritico*S_dif,
           color='r', label='L√≠mites cr√≠ticos') # LS IC

# Fill de 1 - alfa
plt.fill_between(x_m, pdf_m,
                 where = (x_m < -Tcritico*S_dif),
                 color='r', alpha=0.1, label=r'$\alpha/2$')
plt.fill_between(x_m, pdf_m,
                 where = (x_m > Tcritico*S_dif),
                 color='r', alpha=0.1, label=r'$\alpha/2$')
# Fill de 1 - beta (Potencia)
plt.fill_between(x, pdf,
                 where = (x < -Tcritico*S_dif),
                 color='g', alpha=0.1, label=r'$\beta$')
# Tama√±o del efecto
plt.hlines(xmin=0, xmax=xbar, y=pdf.max()+1,  colors='cyan', alpha=0.5, label='Tama√±o del efecto')
plt.vlines(ymin=pdf.max()+0.5, ymax=pdf.max()+1.5, x=0,  colors='cyan', alpha=0.5)
plt.vlines(ymin=pdf.max()+0.5, ymax=pdf.max()+1.5, x=xbar,  colors='cyan', alpha=0.5)
plt.text((xbar/2), pdf.max()+1.5, f'Tama√±o efecto = {dif_bruta:4.2f}', fontsize=12, ha='center')

# Anotaciones
plt.text(xbar, pdf.max()/3, f'Potencia = {1-beta:4.2f}', ha='center')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```
