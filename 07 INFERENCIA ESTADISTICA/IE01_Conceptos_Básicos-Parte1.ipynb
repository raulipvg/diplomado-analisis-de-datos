##<font color='blue'>__Mapa conceptual__</font>

La __estadística__ es una disciplina que se encarga de procesar y organizar los datos, siendo los datos cualquier medida o valor que se puede obtener a través de experimentos, encuestas, censos u otros medios. El análisis de los datos por lo general se inicia con la aplicación de métodos de __estadística descriptiva__, para luego seguir con métodos de __estadística inferencial__.
La primera, se concentra en recolectar, organizar,  analizar e interpretar un conjunto de datos; mientra que la segunda, reune un conjunto de métodos que implican el uso de datos muestrales para hacer generalizaciones o inferencias acerca de una población.
<img alt="Estadistica" src="https://drive.google.com/uc?export=view&id=1YqmXclMiSEo8BE2is7E24dEFZbTGlfuL" width=960px>


Recordemos algunos conceptos:

- __Población:__ Conjunto de elementos en los que se observa alguna característica común, a la cual llamaremos __variable estadística__.

- __Variable estadística__: Es una característica de una muestra o población de datos que puede adoptar diferentes valores.

- __Parámetro:__ Función numérica que describe una variable observada en la población (e.g.: la media ($\mu$) y la varianza ($\sigma^2)$).

- __Muestra__: Conjunto de unidades representativas de una población, seleccionadas de forma aleatoria.

- __Estadístico__: Función numérica de los valores de la muestra (e.g., la media muestral ($\bar{x}$) y la varianza muestral ($S^2$)).

## <font color='blue'>**Conceptos básicos de muestreo**</font>

Un error frecuente sostiene que la era de Big data, significa el fin de la necesidad de **muestreo**. De hecho, la proliferación de datos de diferente calidad y relevancia, refuerza la necesidad del muestreo como herramienta para trabajar de manera eficiente con una variedad de datos y minimizar el sesgo (_bias_). Incluso en un proyecto de Big data, los modelos predictivos suelen desarrollarse con muestras.

### Muestra aleatoria y Sesgo de muestra

Si deseamos poder extraer conclusiones validas sobre la población, la  muestra debe ser **representativa de la población**. ¿Y como logramos esto? Realizando **Muestreo Aleatorio**.

El muestreo aleatorio es un proceso en el que cada unidad de análisis de la población, tiene la misma probabilidad de ser seleccionado para componer la muestra. La muestra resultante se denomina **muestra aleatoria**.

El muestreo aleatorio se puede realizar con y sin reemplazo:

En el **muestreo  con reemplazo**, después de que una unidad de análisis  ha sido seleccionada de entre el marco de la muestra se devuelve y es elegible para ser seleccionado de nuevo.

En el **muestreo  sin reemplazo**, después de que una unidad de análisis  ha sido seleccionada del marco de la muestra, se retira de la población y no regresa a la base del muestreo.

No olvidar que la calidad de los datos a menudo es más importante que la cantidad de ellos cuando se hace una estimación o un modelo basado en una muestra. La calidad de los datos en la ciencia de datos implica la integridad, la coherencia del formato, la limpieza y la precisión de los puntos de datos individuales. La estadística agrega la noción de **representatividad**.

### <font color='blue'>**Sesgo (Bias)**</font>

El __sesgo estadístico__ se refiere a **errores de medición o de muestreo** que son **sistemáticos** y producidos por el proceso de medición o muestreo. Debe hacerse una distinción importante entre los errores debidos al **azar** y los errores debidos **al sesgo**.

Considere el proceso físico de un arma que dispara a un objetivo. La bala no impactará el centro absoluto al apuntar cada vez. Un proceso imparcial producirá un error, pero el mismo será aleatorio y no tenderá a preferenciar ninguna dirección. En la imagen siguiente tenemos a la izquierda un scatter plot sin sesgo y en el gráfico de la derecha uno con sesgo.

### <font color='blue'>**Muestreo Aleatorio**</font>

1. **Muestreo aleatorio simple**: cada unidad de análisis tiene la misma probabilidad de ser seleccionado para ser parte de la muestra. Se utiliza cuando no tenemos ningún tipo de información previa sobre la población objetivo.


<img alt="MuestreoAleatorioSimple" src="https://drive.google.com/uc?export=view&id=15fqiOv35ofZlyBTw36A4GpF0QndDyFkn" width=380px>


2. **Muestreo estratificado**: el objetivo del proceso de muestreo es separar la población en segmentos homogéneos en su interior (estratos), y luego con un muestreo aleatorio simple se seleccionan unidades de análisis de cada segmento (estrato). Las muestras que se obtienen de cada estrato, pueden ser proporcionales o no al tamaño relativo (peso) del estrato. Esto dependerá del tipo de pregunta de investigación que nos hayamos hecho previamente. El problema del muestreo estratificado es que la mayoría de las veces no es sencillo reconocer los estratos y/o su peso en una población.

<img alt="MuestreoEstratificado" src="https://drive.google.com/uc?export=view&id=1CObafZxcMpFBWVNkgsZLiclGC76ev_xV" width=380px>



3. **Muestreo por conglomerados (grupos)**: la población se divide en grupos ( secciones o clusters) y luego se realiza un __muestreo simple__ para seleccionar grupos de forma al azar. Por lo tanto, la unidad de análisis es el grupo.
Las unidades de muestreo o grupos pueden ser espaciados, tal como ocurre naturalmente en las unidades geográficas o físicas (e.g., regiones, comunas); en base a una organización como escuelas, grado escolar; o servicio telefónico tales como códigos de área de los números de teléfono.
La heterogeneidad del grupo debe ser representativa de la heterogeneidad de la población. Todos los elementos del conglomerado se utilizan para el muestreo. Los grupos o clusters se identifican utilizando detalles como edad, sexo, ubicación, etc. Ejemplo: en un colegio existen 4 niveles por curso y 12 cursos (de primero básico a cuarto medio) en total. Un muestreo por conglomerado tomaría un curso por nivel para su muestra, obteniento una muestra de 12 grupos.

<img alt="Muestreo por grupos" src="https://drive.google.com/uc?export=view&id=1Z7gCfP0wLaR_iy-QJ-EPnZPYO2978vti" width=380px>


4. **Muestreo de varias etapas**: Es la combinación de uno o más métodos descritos anteriomente. La población se divide en múltiples grupos (conglomerados), por definición heterogéneos, y luego estos grupos se pueden subdividir y/o agrupar en varios subgrupos (heterogéneos) o estratos (homogéneos). Este proceso continúa según la necesidad de análisis. En cada etapa, se seleccionan unidades de análisis al azar. Ejemplo: los establecimientos educacionales de una región se pueden agrupar por comuna, por tipo (básica, media, etc.), se pueden estratificar por tipo de servicio (particular o público) o nivel (primero básico, segundo básico, etc.).

## <font color='blue'>Teoría de Probabilidad</font>

La probabilidad trata de medir la incertidumbre y consiste en interpretar y comprender los eventos aleatorios de la vida, y dado que vivimos en absoluta aleatoriedad (incluso si tratamos de creer que no lo hacemos), su utilidad se vuelve bastante clara. La probabilidad es la posibilidad a largo plazo de que se produzca un determinado resultado de algún proceso aleatorio. Básicamente, nos dice con qué frecuencia ocurrirán diferentes tipos de eventos.

Algunos ejemplos de la vida real del uso de probabilidades son:

- Finanzas: al estimar la probabilidad de que un activo financiero determinado se encuentre entre o dentro de un rango específico, es posible desarrollar estrategias comerciales para capturar ese resultado previsto.

- Pronóstico del tiempo: los meteorólogos no pueden predecir exactamente qué tiempo hará, por lo que utilizan herramientas e instrumentos para determinar la probabilidad de que llueva, nieve o granizo. También examinan bases de datos históricas para estimar temperaturas altas y bajas, y patrones climáticos probables para ese día o semana.

- Seguro: la probabilidad juega un papel importante en el análisis de las pólizas de seguro para determinar qué planes son los mejores para los clientes y qué deducibles necesitan.

- Deportes: los atletas y entrenadores usan la probabilidad para determinar las mejores estrategias deportivas para juegos y competencias. Empresas como BWin han hecho un negocio con esto e incluso puedes apostar usando diferentes estrategias.

- Publicidad: la probabilidad se utiliza para estimar los clientes potenciales que tendrán más probabilidades de reaccionar positivamente a campañas específicas, en función de sus patrones de consumo.

Veamos el concepto de probabilidad a través del siguiente ejemplo. Definamos nuestro experimento como el lanzamiento de un dado. Entenderemos como:

- __Espacio Muestral__: el conjunto de todos los posibles resultados del experimento. Se denota con la letra $\Omega$

$$\Omega = \{1, 2, 3, 4, 5, 6\}
$$
- __Evento o suceso__: cualquier subconjunto del espacio muestral. Se denota con cualquier letra mayúscula. Ejemplo: sea $A$: lanzar el dado y obtener un número par
$$
A = \{2, 4, 6\}
$$

- __Regla de Laplace__: cuando un experimento aleatorio es regular, es decir, que todos los eventos tiene la misma probabilidad de ocurrir (equiprobables), entonces para calcular la probabilidad de un suceso cualquiera $A$, se obtiene a partir del cociente entre la cantidad de elementos de $A$ divididos por el número de elementos de espacio muestral $\Omega$ (casos totales).
<br><br>

$$
P(A) = \frac{\text{número de casos favorables}}{\text{número de casos totales}}
$$
<br>

En nuestro ejemplo, la probabilidad del evento $A$ sería:

$$
P(A) = \frac{3}{6} = 0.5
$$



La probabilidad de un evento es un número que va desde 0 (lo que significa que no hay forma de que suceda un evento) a 1 (lo que significa que el evento sucederá con seguridad). La suma de la probabilidad de todos los resultados posibles y los suma da 1. Cuanto mayor sea el valor de la probabilidad, mayor será la probabilidad de que ocurra el evento.

Axiomas:

1. La probabilidad de un evento ${\displaystyle S}$ no puede ser negativa

$$0 \le P({\displaystyle S})$$

2. La probabilidad del evento seguro, ${\displaystyle \Omega }$ , es igual a 1, denotado simbólicamente como:

$$1 = P({\displaystyle \Omega})$$

3. Si ${\displaystyle E_{1},E_{2},\dots }$ son eventos mutuamente excluyentes (es decir, no tienen resultados en común o su intersección es el conjunto vacío), entonces:

$$ P(\displaystyle E_{1} \cup \displaystyle  E_{2}\cup \dots )=\sum P(E_{i})$$

Experimento: Se echa a rodar un dado de 6 lados. ¿Cuál es la probabilidad de que salga un número par o uno impar?

Posibilidades:
1. El número que salga puede ser un número impar.
2. El número que salga puede ser un número par.

Eventos:

Estos eventos son mutuamente excluyentes no pueden ocurrir a la vez.




### <font color='blue'>**Variables aleatorias**</font>

Digamos que la probabilidad de que llueva mañana es del 40% (expresada como "P (lluvia) = 0,4"), o la probabilidad de un robo de automóvil en una región en particular es del 2% (definida como " P (robo de coche) = 0,02 ”). En el primer caso, le interesa la variable "lluvia", y en el segundo, la variable "robo de coche". Estas variables, así como cualquier otra que sea resultado de un proceso aleatorio, se denominan **variables aleatorias**.

Una **variable aleatoria** es una función matemática de un experimento aleatorio. En otras palabras, es una ecuación que describe o intenta describir los resultados (con un número) de un evento cuyos resultados se deben al azar.



### <font color='blue'>**Distribuciones de probabilidad**</font>

Una distribución de probabilidad es una lista de todos los posibles resultados de una variable aleatoria, junto con sus correspondientes valores de probabilidad. Una distribución de probabilidad vincula cada resultado de una variable o proceso aleatorio con su probabilidad de ocurrencia. Por ejemplo, si toma muestras de sangre de diferentes personas en un lugar determinado, puede calcular la distribución de probabilidad de sus tipos de sangre:

<img alt="Distribución de probabilidad" src="https://drive.google.com/uc?export=view&id=10rCuuQAnuMleOejdEVnaOOrUI5nYnuqz" width=480px>


En este caso, el tipo de sangre es la variable aleatoria. La distribución de probabilidad muestra que los individuos con el tipo de sangre "A" tienen la mayor probabilidad de ocurrencia, a diferencia de las personas con el tipo de sangre "AB", que tienen la más baja. Además, si sumas las diferentes probabilidades, obtienes 1. ¿Cómo lo haces? Simple: define la variable aleatoria que desea analizar (en este ejemplo es "tipo de sangre"), establece las categorías que desea contar ("O", "A", "B" y "AB"), y cuente el número de ocurrencias en cada categoría. Luego, o trabaja con ese número absoluto (por ejemplo, 150 casos de tipo de sangre "A"), o lo divide por el número total de casos (haciéndolo un porcentaje del total, como en nuestro ejemplo).

Lo interesante es cuando **estima** la distribución de probabilidad de una variable aleatoria (como el tipo de sangre o cualquier otra), lo que realmente está haciendo es usar datos que representan solo una parte del comportamiento real de esa variable aleatoria que está analizando. No está mirando todos los valores de datos posibles (que en términos probabilísticos se define como __población__), ya que solo obtuvo datos de un subconjunto (que se conoce como __muestra__) en un momento y espacio dados.

```python
# Tu código aquí ...
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = '/content/drive/MyDrive/07_Inferencia_Estadistica/loans_income.csv'
df = pd.read_csv(file)
```

```python
#tamaño de la poblacion
len(df)
```

```python
samples = [10, 50, 100, 1000, 10000, 50000]
dstats = {}

for n in samples:
  df_n = df.sample(n).describe().T.reset_index().drop(columns='index')
  d = {}
  for col in df_n:
    d[col] = df_n.loc[0,col]
  dstats[n] = d
stats_final = pd.DataFrame(dstats)
stats_final
```

```python
# Crear el gráfico
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")

# Graficar cada estadístico
estadisticos = ['mean', 'std', '25%', '50%', '75%']
nsample = stats_final.columns.to_list()
for i in estadisticos:
    plt.semilogx(nsample, stats_final.loc[i], marker='o', label=i)

# Personalizar el gráfico
plt.title("Comparación de Estadísticos entre Muestras y Población", fontsize=14)
plt.xlabel("Tamaño de Muestra", fontsize=12)
plt.ylabel("Valor del Estadístico", fontsize=12)
plt.legend(title="Estadístico", fontsize=10)
plt.grid(True)
plt.show()
```

A medida que las muestras son de mayor tamaño podemos ver que sus estadísticos como media, mediana, desviación estandar y rango intercuartil convergen al valor de la población (tamaño población = 50.000)
